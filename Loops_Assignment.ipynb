## Student Marks Entry

### Goal:
    Record student marks and skip invalid entries.

### Steps:

    Create an empty list marks = [].
    
    Ask user to enter marks for 5 students.
    
    If marks < 0 or > 100 → print “Invalid marks” and continue.
    
    If user enters -1 → stop taking input (break).
    
    Print valid marks and their average.

### Example:

    Enter marks: 90  
    Enter marks: 105 → Invalid marks  
    Enter marks: 80  
    Enter marks: -1 → stop  
    Valid Marks: [90, 80]  
    Average = 85.0

```python
marks=[]

while True:
    input_user=int(input("Enter Marks"))

    if input_user == -1:
        break
    if input_user < 0 or input_user > 100:
        continue
    marks.append(input_user)
        
print("Valid Marks",marks)
Average=sum(marks) / len(marks) 
print(Average)

    
```

## 🏨 8. Hotel Room Booking

### Goal:
    Manage simple room booking.

### Task:

    Create a list of rooms [101, 102, 103, 104, 105]
    
    Ask user to choose a room.
    
    If room already booked, print “Room not available.”
    
    Continue until all rooms are booked.

### Example:

    Available rooms: [101, 102, 103, 104, 105]
    Choose room: 103
    Room 103 booked successfully!

```python
room=[101, 102, 103, 104, 105]
book_room=[101,104,105]

print('Rooms: ',room)
while True:
    
    user_i=int(input('Choose room: '))
    if user_i in book_room:
        print("Room not available.")
        continue
    else:
        book_room.append(user_i)
        print(f'Room {user_i} booked successfully!')
        break

print('Booked Room ',book_room)
```

## Invoice Tax Calculator
